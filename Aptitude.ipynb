**VERBAL ABILITY ERROR CORRECTION TYPES**

**1. Redundant Contrasting Conjunctions and Related Errors**
example question: "Even though it was raining, but we still went for a walk."
answer is and reason: Correct to "Even though it was raining, we still went for a walk." or "It was raining, but we still went for a walk." "Even though" and "but" both signal contrast—using both repeats the same function and is redundant.

**2. Misuse of "despite" and "in spite of"**
example question: "Despite of the traffic, we arrived on time."
answer is and reason: Correct to "Despite the traffic, we arrived on time." or "In spite of the traffic, we arrived on time." "Despite" already includes "of," so adding "of" is incorrect.

**3. Redundant Use of "because" and "reason"**
example question: "The reason I'm late is because my car broke down."
answer is and reason: Correct to "The reason I'm late is that my car broke down." or "I'm late because my car broke down." Don't pair "reason" + "because"—either use "reason… is that" or simply "because."

**4. Double Comparatives or Superlatives**
example question: "This is more better than I expected."
answer is and reason: Correct to "This is better than I expected." You can't use "more" with an already comparative adjective.

**5. Misplaced "only"**
example question: "I only eat vegetables." (meaning you eat nothing else)
answer is and reason: Correct to "I eat only vegetables." "Only" must immediately precede the word or phrase it modifies.

**6. Redundant Prepositions**
example question: "Where are you at?"
answer is and reason: Correct to "Where are you?" The preposition "at" is unnecessary—"where" already expresses location.

**7. Misuse of "like" and "as"**
example question: "Do it like I showed you."
answer is and reason: Correct to "Do it as I showed you." Use "as" to introduce clauses, "like" only before nouns or pronoun + noun.

**8. Faulty Parallelism with Contrasting Ideas**
example question: "The movie was long but also it was boring."
answer is and reason: Correct to "The movie was long but also boring." Keep coordinate elements parallel by omitting the repeated subject + verb.

**9. Misplaced Modifiers with Contrasting Ideas**
example question: "Although broken, she still loved the antique vase."
answer is and reason: Correct to "Although the vase was broken, she still loved it." The modifier must clearly refer to the vase, not "she."

**10. Redundant Negatives in Contrasting Statements**
example question: "She didn't say nothing, but she looked angry."
answer is and reason: Correct to "She didn't say anything, but she looked angry." Two negatives cancel out—use a single negative + "anything."

**11. Redundancy in Verbs + Prepositions ("discuss about," "awaiting for")**
example question: "We will discuss about the project tomorrow."
answer is and reason: "We will discuss the project tomorrow." The verb "discuss" already takes a direct object—adding "about" is redundant.
example question: "The students were awaiting for the chief guest."
answer is and reason: "The students were awaiting the chief guest." The verb "await" does not require "for."

**12. Wrong Verb Choice for Animate vs. Inanimate ("got drowned" vs. "sink")**
example question: "The ship got drowned in the storm."
answer is and reason: "The ship sank in the storm." Only living beings "drown"; ships and objects "sink."

**13. Correct Preposition in Phrasal Verb ("put up at" vs. "put up in")**
example question: "We put up in a hotel last night."
answer is and reason: "We put up at a hotel last night." Use "at" for lodging at a general location; "in" for specific enclosed spaces.

**14. Conditional Tense Sequence Errors**
example question: "If I had known this yesterday, I will have helped him."
answer is and reason: "If I had known this yesterday, I would have helped him." Third conditional needs "would have + past participle."

**15. Prepositional Distinctions ("on behalf of" vs. "by")**
example question: "She spoke by the committee."
answer is and reason: "She spoke on behalf of the committee." "On behalf of" means as representative; "by" introduces the agent in passive voice.

**16. Incorrect Tense for Specific Past Events**
example question: "I have got my M.Sc. degree in 1988."
answer is and reason: "I got my M.Sc. degree in 1988." Use simple past for actions at a definite past time, not present perfect.

**17. Missing or Wrong Article with Adjectives**
example question: "It is meek and humble who shall inherit the earth."
answer is and reason: "It is those who are meek and humble who shall inherit the earth." Or "It is the meek and the humble …" You need "those" or "the" to make it grammatical.

**18. Adjective vs. Adverb and Comparative Forms**
example question: "The roses smell more sweetly than I thought."
answer is and reason: "The roses smell sweeter than I thought." Use comparative adjective "sweeter" to compare a quality of the roses.

**19. Subject-Verb Agreement & Capitalization**
example question: "Block of Residential flats are coming up near our house."
answer is and reason: "A block of residential flats is coming up near our house." Singular subject "block" takes "is," and mid-sentence "residential" is not capitalized.

**20. Redundancy in Relative Clauses**
example question: "You can get all the information that you want in this book."
answer is and reason: "You can get all the information you want in this book." The relative pronoun "that" is removable when it adds no new information.

**21. Quantity as Singular Subjects**
example question: "Sixty miles are a good distance."
answer is and reason: "Sixty miles is a good distance." Measurements take singular verbs when treated as a single quantity.

**22. Redundant Intensifiers ("at great length")**
example question: "He explained the matter at great length…"
answer is and reason: "He explained the matter at length…" "Great" is redundant—"at length" already means in detail.

**23. Idiomatic Expressions & Article Use ("come into contact")**
example question: "They came into the contact with many experts."
answer is and reason: "They came into contact with many experts." "Contact" is uncountable here—omit "the."

**24. Collective Noun Terminology**
example question: "A flock of cows grazed in the field."
answer is and reason: "A herd of cows grazed in the field." Use "herd" for large herbivores, "flock" for birds or sheep.

**25. Precise Word Choice in Context ("restoration" vs. "resurrection")**
example question: "The resurrection of old paintings is delicate work."
answer is and reason: "The restoration of old paintings is delicate work." "Restoration" means repairing to original state; "resurrection" means bringing back to life.

**26. Expressing Non-necessity Naturally**
example question: "You need not come unless you want to."
answer is and reason: "You don't need to come unless you want to." "Don't need to" is more idiomatic in modern usage.

**27. Parallel Structure in Habitual Actions**
example question: "The small child does whatever his father did."
answer is and reason: "The small child does whatever his father does." Keep the verb tense parallel when describing habitual actions.

**28. Phrasal-Verb + Preposition ("join in crushing")**
example question: "The leaders joined the princes at crushing the serfs."
answer is and reason: "The leaders joined the princes in crushing the serfs." Use "join in [verb]-ing" for participation.

**29. Records and "break" vs. "meet" vs. "better"**
example question: "The record has not been met since 1911."
answer is and reason: "The record has not been broken since 1911." "Break a record" is the correct idiom.

**30. Inverted Third Conditional**
example question: "If the room had been brighter, I would have been able to read."
answer is and reason: "Had the room been brighter, I would have been able to read." Inversion omits "if" and moves "had."

**31. Phrasal-Verb Meaning Distinctions ("look after" vs. "look up")**
example question: "She will look up the marketing network."
answer is and reason: "She will look after the marketing network." "Look after" means manage or take care of; "look up" means search for information.

**32. Plural Subjects with Plural Verbs**
example question: "The Indians in general is not vigilant."
answer is and reason: "The Indians in general are not vigilant." Plural "Indians" requires plural verb "are."

**33. Tense Consistency in Complex Sentences**
example question: "She didn't go because she has to work."
answer is and reason: "She didn't go because she had to work." Past context needs past tense "had."

**34. Neutral Quantity Expressions ("so many" vs. "too many")**
example question: "I have read too many books by him."
answer is and reason: "I have read so many books by him." "Too many" implies excessive (negative), "so many" simply large in number.

**35. Common Confusable Words**
(For each pair below, the first sentence shows the error; the correction uses the right form.)
• Affect/effect: "The movie effected me deeply." → "The movie affected me deeply."
• Except/accept: "Everyone excepted the gift." → "Everyone accepted the gift."
• Lose/loose: "Don't loose your keys." → "Don't lose your keys."
• They're/their/there: "There going home." → "They're going home."
• Two/to/too: "I have to apples." → "I have two apples."
• Data/datum: "This datum are incorrect." → "This data are incorrect."
• Each other's: "They admired each others' work." → "They admired each other's work."
• A lot/alot: "There's alot to do." → "There's a lot to do."
• Then/than: "Better then nothing." → "Better than nothing."
• Ensure/insure: "I insure your safety." → "I ensure your safety."
• Principle/principal: "The principal reason…" → "The principle reason…" (if meaning rule) or swap context.
• Fewer/less: "Less people came." → "Fewer people came."
• Farther/further: "We'll discuss this farther." → "We'll discuss this further."
• Lay/lie: "I am going to lay down." → "I am going to lie down."
• Die/dye: "Plants will dye without water." → "Plants will die without water."
• Discreet/discrete: "Keep it discrete." → "Keep it discreet."
• Elicit/illicit: "They were arrested for eliciting drugs." → "They were arrested for illicit drugs."
• Precede/proceed: "We will precede with the plan." → "We will proceed with the plan."
• A/an: "She is a honest person." → "She is an honest person."
• Advice/advise: "Please advice me." → "Please advise me."
• Compliment/complement: "That sauce will compliment the steak." → "That sauce will complement the steak."
• Latter/ladder: "I prefer the ladder option." → "I prefer the latter option."
• Lead/led: "She led the team." (correct past), "Who will lead?" (present)
• Alley/ally: "The cat ran down the ally." → "The cat ran down the alley."
• Ensure/assure: "I assure the doors are locked." → "I ensure the doors are locked."
