# spark: spark is a engine framework with technologies like :
- spark-sql
- spark-

| Feature              | Hive (on MapReduce)                  | Spark SQL                             |
| -------------------- | ------------------------------------ | ------------------------------------- |
| **Engine**           | Uses **MapReduce**                   | Uses **Spark engine** (in-memory)     |
| **Speed**            | **Slower** (reads/writes to disk)    | **Faster** (in-memory processing)     |
| **Latency**          | High (batch only)                    | Low (supports real-time)              |
| **Language Support** | Mostly **SQL (HiveQL)**              | **SQL**, **Scala**, **Python**, etc.  |
| **Use Case**         | Good for **batch processing**        | Good for **batch + streaming**        |
| **Execution**        | Writes intermediate data to **disk** | Keeps intermediate data **in memory** |

In short:

* Use **Hive** if you're okay with slow, batch jobs.
* Use **Spark** if you want fast, flexible, and modern data processing.


**PySpark** is the **Python API** for **Apache Spark**, which is originally written in **Scala**.

So:

* Spark is built in **Scala**
* PySpark **wraps** the Scala Spark engine using **Py4J** (a bridge between Python and JVM)
* When you write PySpark code, it runs on the **same Spark engine**, just via Python

You're using the **same core Spark**, just controlling it with Python instead of Scala.
