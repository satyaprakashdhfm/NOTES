````javascript
function reverseDigit(num){
    let reversed =0;
    while(num > 0){
        let digit = num % 10;
        reversed = reversed * 10 + digit;
        num = Math.floor(num / 10);
    }
    return reversed;
}

function isPalindrome(string){
    let reverse = '';
    for (let i = string.length-1;i>=0;i--) {
        reverse = reverse + string[i];
    }
    return reverse==string ;
}

function reverserString(string){
    let reversed = '';
    for (let i = string.length - 1; i >= 0; i--) {
        reversed += string[i];
    }
    return reversed;
}

function sqrt(num) {
    return Math.sqrt(num);
}

function isPrime(num) {
    if (num <= 1) return false;
    for (let i = 2; i <= Math.sqrt(num); i++) {
        if (num % i === 0) return false;
    }
    return true;
}

function gcd(a,b){
    while (b!=0){
        let temp = b;
        b= a % b;
        a = temp;
    }
    return a;
}

function lcm(a, b) {
    return (a * b) / gcd(a, b);
}

function factorial(n) {
    if (n === 0 || n === 1) return 1;
    let result = 1;
    for (let i = 2; i <= n; i++) {
        result *= i;
    }
    return result;
}


console.log(reverseDigit(12345)); // Output: 54321
console.log(isPalindrome("racecar")); // Output: true
console.log(reverserString("hello")); // Output: "olleh"
console.log(sqrt(16)); // Output: 4
console.log(isPrime(7)); // Output: true
console.log(gcd(48, 18)); // Output: 6
console.log(lcm(4, 5)); // Output: 20
console.log(factorial(5)); // Output: 120




````

###  What is **Hoisting** in JavaScript?

ðŸ”Ž How Hoisting Works (with `var`)

```js
console.log(x); // undefined
var x = 5;
```

âœ… JavaScript behind the scenes:

```js
var x;          // hoisted
console.log(x); // undefined
x = 5;
```

* `var x` is **hoisted** (declared at top), but the **value assignment stays in place**.
* So before line 2, `x` exists but is `undefined`.

### `let` and `const` are **not** hoisted the same way

```js
console.log(y); // âŒ ReferenceError
let y = 10;
```

* `let` and `const` **are hoisted**, but **not initialized**.
* This causes a **"temporal dead zone"** where using the variable before its declaration gives an error.

### Function Declarations ARE hoisted

```js
sayHello(); // âœ… Works

function sayHello() {
  console.log("Hello!");
}
```

* Entire function is hoisted â€” so you can call it even before its definition.

### Function Expressions are NOT hoisted

```js
greet(); // âŒ TypeError: greet is not a function

var greet = function() {
  console.log("Hi!");
};
```

* `var greet` is hoisted (as `undefined`), but the function itself is **not assigned yet**.

````javascript
// classes and objects
// method and variable types
class Person {
  // Class variable (shared across all instances)
  static numberOfPeople = 0;

  constructor(name, age) {
    this.name = name; // Instance variable
    this.age = age;   // Instance variable
    Person.numberOfPeople++; // Increment class (static) variable
  }

  // Instance Method
  greet() {
    return `Hello, my name is ${this.name} and I am ${this.age} years old.`;
  }

  // Class Method (static)
  static getNumberOfPeople() {
    return Person.numberOfPeople;
  }

  // Static Method
  static isAdult(age) {
    return age >= 18;
  }
}

// Creating instances
const person1 = new Person("Alice", 30);
const person2 = new Person("Bob", 25);

// Using Instance Method
console.log(person1.greet()); // Hello, my name is Alice and I am 30 years old.

// Using Class Method
console.log(Person.getNumberOfPeople()); // 2

// Using Static Method
console.log(Person.isAdult(30)); // true
console.log(Person.isAdult(16)); // false

// Calling static method from instance (possible but not ideal)
console.log(person1.constructor.isAdult(30)); // true

// Print method references
console.log(Person.getNumberOfPeople); // [Function: getNumberOfPeople]
console.log(Person.isAdult);           // [Function: isAdult]
````

````javascript
// âœ… 1. Single Inheritance
class Computer {
  constructor() {
    this.brand = ''; // instance variable
  }

  displayBrand() {
    console.log("Brand: " + this.brand);
  }
}

class Laptop extends Computer {
  constructor() {
    super();
    this.batteryLife = 0;
  }

  displayBatteryLife() {
    console.log("Battery Life: " + this.batteryLife + " hours");
  }
}

const myLaptop = new Laptop();
myLaptop.brand = "Dell";
myLaptop.batteryLife = 10;
myLaptop.displayBrand();           // Output: Brand: Dell
myLaptop.displayBatteryLife();    // Output: Battery Life: 10 hours


// âœ… 2. Multi-level Inheritance
class Animal {
  eat() {
    console.log("Animal is eating");
  }
}

class Dog extends Animal {
  bark() {
    console.log("Dog is barking");
  }
}

class Puppy extends Dog {
  weep() {
    console.log("Puppy is weeping");
  }
}

const myPuppy = new Puppy();
myPuppy.eat();   // inherited from Animal
myPuppy.bark();  // inherited from Dog
myPuppy.weep();  // own method


// âœ… 3. Hierarchical Inheritance
class Vehicle {
  start() {
    console.log("Vehicle is starting");
  }
}

class Car extends Vehicle {
  drive() {
    console.log("Car is driving");
  }
}

class Bike extends Vehicle {
  ride() {
    console.log("Bike is riding");
  }
}

const myCar = new Car();
myCar.start();  // from Vehicle
myCar.drive();  // own method

const myBike = new Bike();
myBike.start(); // from Vehicle
myBike.ride();  // own method


// âœ… 4. Multiple Inheritance via Mixins (simulate interface-style behavior)
const AnimalSound = {
  sound() {
    console.log("Animal makes sound");
  }
};

const PetSound = {
  sound() {
    console.log("Pet makes sound");
  }
};

// Mixing both interfaces
class DogWithMix {
  sound() {
    console.log("Dog makes a sound");
  }
}

// Apply mixins (simulate multiple interfaces)
Object.assign(DogWithMix.prototype, AnimalSound, PetSound);

const myDog = new DogWithMix();
myDog.sound();  // Calls overridden DogWithMix method
````

### **Polymorphism in JavaScript**

### **Definition**

Polymorphism is the ability of different objects to respond to the same function or method call in their own way. It's about having **one interface, many implementations**.

> JavaScript is **dynamically typed** and **interpreted**, so most polymorphism in JS is **runtime-based** â€” like Python.



## âœ… **Types of Polymorphism in JavaScript**

 1. **Runtime Polymorphism**

> Happens when the method to invoke is determined at **runtime**

âœ… a. **Method Overriding (Inheritance-based)**

```js
class Animal {
  speak() {
    console.log("Animal makes a sound");
  }
}

class Dog extends Animal {
  speak() {
    console.log("Dog barks");
  }
}

class Cat extends Animal {
  speak() {
    console.log("Cat meows");
  }
}

const a1 = new Dog();
const a2 = new Cat();

a1.speak(); // Dog barks
a2.speak(); // Cat meows
```

âœ… b. **Duck Typing (Like Python)**

> JS doesn't care what type an object is â€” if it has the method, it works.

```js
function makeItSpeak(animal) {
  animal.speak();
}

const bird = {
  speak: () => console.log("Bird sings")
};

makeItSpeak(bird);  // Bird sings
```

ðŸ”¹ 2. **Compile-Time Polymorphism (Simulated Only)**

> JavaScript does **not support** real compile-time polymorphism like C++ or Java.

âš ï¸ a. **Method Overloading â€” Not Supported Natively**

> You **can't define multiple methods** with the same name and different parameters.

```js
class Greet {
  // Only one definition allowed; last one will overwrite
  sayHello(name) {
    console.log("Hello " + name);
  }

  // This will overwrite the previous sayHello()
  // sayHello(name, age) { ... } âŒ Not valid overloading
}
```

âœ… Simulated Using **Default/Rest Parameters**

```js
class Greet {
  sayHello(name = "Guest", age = null) {
    if (age !== null) {
      console.log(`Hello ${name}, age ${age}`);
    } else {
      console.log(`Hello ${name}`);
    }
  }
}

const g = new Greet();
g.sayHello();           // Hello Guest
g.sayHello("Alice");    // Hello Alice
g.sayHello("Bob", 25);  // Hello Bob, age 25
```

ðŸ”¹ 3. **Operator Overloading**

> âŒ **Not supported** in JS
> JS does not allow operator overloading directly (e.g., redefining `+` for objects), unlike Python or C++.

```js
// Example that would work in Python but not in JS:
// obj1 + obj2  âŒ Invalid in JS
```

````javascript
// Encapsulation:

class BankAccount {
  // âœ… Private field (true private in JS using #)
  #accountNumber;

  // âœ… Public fields
  accountHolder;
  branchName;

  // âš ï¸ "Protected" simulation (by convention using underscore _balance)
  _balance;

  // âœ… Constructor
  constructor(accountNumber, accountHolder, balance, branchName) {
    this.#accountNumber = accountNumber;
    this.accountHolder = accountHolder;
    this._balance = balance;
    this.branchName = branchName;
  }

  // âœ… Private method (accessible only within the class)
  #showPrivateInfo() {
    console.log(`Private Info: ${this.#accountNumber}`);
  }

  // âœ… Public method: can access private method internally
  getAccountNumber() {
    this.#showPrivateInfo(); // Internal access
    return this.#accountNumber;
  }

  // âœ… Public getter & setter
  getAccountHolder() {
    return this.accountHolder;
  }

  setAccountHolder(name) {
    this.accountHolder = name;
  }

  // âœ… Public read-only getter for balance
  getBalance() {
    return this._balance;
  }

  // âœ… Public method with validation
  deposit(amount) {
    if (amount > 0) {
      this._balance += amount;
      console.log(`Deposited: ${amount}`);
    } else {
      console.log("Invalid deposit amount!");
    }
  }

  withdraw(amount) {
    if (amount > 0 && amount <= this._balance) {
      this._balance -= amount;
      console.log(`Withdrawn: ${amount}`);
    } else {
      console.log("Invalid or insufficient funds!");
    }
  }

  // âš ï¸ "Protected" method (simulated with _ prefix)
  _showProtectedInfo() {
    console.log(`Protected Info: Balance = ${this._balance}`);
  }

  // Default/public method (accessible anywhere)
  showBranch() {
    console.log(`Branch: ${this.branchName}`);
  }
}
````

| **Access Type**       | **How to Define**                     | **How to Use**                          | **Notes**                               |
| --------------------- | ------------------------------------- | --------------------------------------- | --------------------------------------- |
| **Private Field**     | `#fieldName`                          | Only inside class via `this.#fieldName` | ðŸ”’ True privacy (ES2022+)               |
| **Public Field**      | `this.fieldName` or declared normally | `object.fieldName`                      | ðŸŒ Accessible from outside              |
| **"Protected" Field** | `this._fieldName` (convention)        | `object._fieldName`                     | âš ï¸ Not enforced, by naming convention   |
| **Private Method**    | `#methodName()`                       | Only within class: `this.#methodName()` | ðŸ”’ Same as private field                |
| **Public Method**     | `methodName()`                        | `object.methodName()`                   | âœ… Standard method                       |
| **Getter/Setter**     | `get name()` / `set name(val)`        | `object.name`, `object.name = "X"`      | ðŸ§  Looks like a field, acts like method |


````javascript
class Animal {
    constructor(name, species) {
        this.name = name;
        this.species = species;
    }
    
    // Instance method
    speak() {
        console.log(`${this.name} makes a sound`);
    }
    
    // Static method
    static getSpeciesCount() {
        return "Many species exist";
    }
    
    // Getter
    get info() {
        return `${this.name} is a ${this.species}`;
    }
    
    // Setter
    set name(newName) {
        this._name = newName;
    }
}

// Inheritance
class Dog extends Animal {
    constructor(name, breed) {
        super(name, "dog"); // Call parent constructor
        this.breed = breed;
    }
    
    // Override method
    speak() {
        console.log(`${this.name} barks`);
    }
    
    // New method
    wagTail() {
        console.log(`${this.name} is wagging tail`);
    }
}

const myDog = new Dog("Buddy", "Golden Retriever");
myDog.speak(); // "Buddy barks"
myDog.wagTail(); // "Buddy is wagging tail"


//Prototypal Inheritance (Traditional)

// Constructor function
function Person(name, age) {
    this.name = name;
    this.age = age;
}

// Adding methods to prototype
Person.prototype.greet = function() {
    console.log(`Hello, I'm ${this.name}`);
};

Person.prototype.getAge = function() {
    return this.age;
};

// Creating instances
const person1 = new Person("Alice", 25);
const person2 = new Person("Bob", 30);

person1.greet(); // "Hello, I'm Alice"

// Inheritance with prototypes
function Student(name, age, grade) {
    Person.call(this, name, age); // Call parent constructor
    this.grade = grade;
}

// Set up inheritance
Student.prototype = Object.create(Person.prototype);
Student.prototype.constructor = Student;

// Add student-specific methods
Student.prototype.study = function() {
    console.log(`${this.name} is studying`);
};

const student1 = new Student("Charlie", 20, "A");
student1.greet(); // "Hello, I'm Charlie"
student1.study(); // "Charlie is studying"
````

````javascript
class Shape {
    area() {
        throw new Error("Area method must be implemented");
    }
    
    perimeter() {
        throw new Error("Perimeter method must be implemented");
    }
}

class Rectangle extends Shape {
    constructor(width, height) {
        super();
        this.width = width;
        this.height = height;
    }
    
    area() {
        return this.width * this.height;
    }
    
    perimeter() {
        return 2 * (this.width + this.height);
    }
}

class Circle extends Shape {
    constructor(radius) {
        super();
        this.radius = radius;
    }
    
    area() {
        return Math.PI * this.radius * this.radius;
    }
    
    perimeter() {
        return 2 * Math.PI * this.radius;
    }
}

// Polymorphic usage
const shapes = [
    new Rectangle(5, 10),
    new Circle(3),
    new Rectangle(3, 4)
];

shapes.forEach(shape => {
    console.log(`Area: ${shape.area()}`);
    console.log(`Perimeter: ${shape.perimeter()}`);
});
````

### Task Queue Management

// Task queue with priority
class TaskQueue {
    constructor() {
        this.tasks = [];
        this.running = false;
    }
    
    add(task, priority = 0) {
        const taskWrapper = { task, priority };
        
        // Insert task based on priority
        const insertIndex = this.tasks.findIndex(t => t.priority < priority);
        if (insertIndex === -1) {
            this.tasks.push(taskWrapper);
        } else {
            this.tasks.splice(insertIndex, 0, taskWrapper);
        }
        
        this.process();
    }
    
    async process() {
        if (this.running || this.tasks.length === 0) return;
        
        this.running = true;
        
        while (this.tasks.length > 0) {
            const { task } = this.tasks.shift();
            try {
                await task();
            } catch (error) {
                console.error('Task failed:', error);
            }
        }
        
        this.running = false;
    }
}

// Usage

taskQueue.add(() => console.log('High priority task'), 10);

// Memoization
function memoize(fn) {
    const cache = new Map();
    
    return function(...args) {
        const key = JSON.stringify(args);
        
        if (cache.has(key)) {
            return cache.get(key);
        }
        
        const result = fn.apply(this, args);
        cache.set(key, result);
        return result;
    };
}

// Usage
const fibonacci = memoize(function(n) {
    if (n <= 1) return n;
    return fibonacci(n - 1) + fibonacci(n - 2);
});

// Lazy evaluation
class LazyArray {
    constructor(generator) {
        this.generator = generator;
    }
    
    map(fn) {
        const generator = this.generator;
        return new LazyArray(function* () {
            for (const item of generator()) {
                yield fn(item);
            }
        });
    }
    
    filter(predicate) {
        const generator = this.generator;
        return new LazyArray(function* () {
            for (const item of generator()) {
                if (predicate(item)) {
                    yield item;
                }
            }
        });
    }
    
    take(n) {
        const generator = this.generator;
        return new LazyArray(function* () {
            let count = 0;
            for (const item of generator()) {
                if (count >= n) break;
                yield item;
                count++;
            }
        });
    }
    
    toArray() {
        return [...this.generator()];
    }
}

// Usage
const lazyNumbers = new LazyArray(function* () {
    for (let i = 1; i <= 1000000; i++) {
        yield i;
    }
});

const result = lazyNumbers
    .filter(n => n % 2 === 0)
    .map(n => n * 2)
    .take(5)
    .toArray();

console.log(result); // [4, 8, 12, 16, 20]
